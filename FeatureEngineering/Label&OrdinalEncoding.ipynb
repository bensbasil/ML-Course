```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder=LabelEncoder()
```

```python
import pandas as pd
df=pd.DataFrame({
    'color': ['red','blue','green','green','red','blue']
})
```

```python
lbl_encoder.fit_transform(df['color'])
```

```python
lbl_encoder.transform(['red'])
```

```python
lbl_encoder.transform(['blue'])
```

```python
lbl_encoder.transform(['green'])
```

## Ordinal Encoding
## encode categorical data intinsic order or ranking. Each category is assigned a numerical value based on its position in the order.
## If we have categorical variable "educational level" with four possible values (high school, college, graudate, post-graduate), we can represent it using oridinal as follows:

1. HIGH SCHOOL: 1
2. COLLEGE: 2
3. GRADUATE: 3
4. POST - GRADUATE: 4

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({
    'size':['small','medium','large','medium','small','large']
})
```

```python
df
```

```python
##create an instance of ordinal encoder and perform fit_transform
encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```
