```python
num_feat=[feat for feat in df_copy.columns if df_copy[feat].dtype!='O']
cat_feat=[feat for feat in df_copy.columns if df_copy[feat].dtype=='O']
```

```python
## Proportion of count data on numerical columns
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features',fontsize=20,fontweight='bold',alpha=0.8,y=1.)

for i in range(0, len(num_feat)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df_copy[num_feat[i]],fill=True,color='r')
    plt.xlabel(num_feat[i])
    plt.tight_layout()
```

## Top 10 App Categories

```python
# 1. Get value counts and reset index to create a proper DataFrame
category = df['Category'].value_counts().reset_index()

# 2. Rename the columns explicitly
category.columns = ['Category', 'Count']

# 3. Plotting
plt.figure(figsize=(15,6))
# Note: Added 'hue' to avoid the palette warning and 'legend=False'
sns.barplot(
    x='Category', 
    y='Count', 
    data=category[:10], 
    hue='Category', 
    palette='hls', 
    legend=False
)

plt.title('Top 10 App Categories')
plt.xticks(rotation=90)
plt.show()
```
