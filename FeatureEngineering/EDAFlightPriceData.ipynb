```python
##Feature Engineering
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arr_Hr']=df['Arrival_Time'].str.split(':').str[0]
df['Arr_Mn']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Arr_Hr']=df['Arr_Hr'].astype(int)
df['Arr_Mn']=df['Arr_Mn'].astype(int)
```

```python
df.info()
```

```python
df['Dep_Hr']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Mn']=df['Dep_Time'].str.split(':').str[1]
df.head()
```

```python
df.info()
```

```python
df['Dep_Hr']=df['Dep_Hr'].astype(int)
df['Dep_Mn']=df['Dep_Mn'].astype(int)
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.info()
df.head(5)
```

```python
def extract_hours(duration):
    if 'h' in duration:
        # Split by 'h' and take the part before it
        return int(duration.split('h')[0])
    else:
        # If there is no 'h', there are 0 hours
        return 0

df['Dur_Hrs'] = df['Duration'].apply(extract_hours).astype(int)
df['Dur_Mn'] = df['Duration'].str.split(' ').str[1].str.replace('m', '').fillna('0').astype(int)
```

```python
df.drop('Duration',axis=1,inplace=True)
df.info()
```
