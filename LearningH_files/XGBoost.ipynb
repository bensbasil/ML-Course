```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import xgboost as  xgb
from sklearn.calibration import LabelEncoder
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE
from sklearn.utils.class_weight import compute_class_weight
from collections import Counter
import pickle
```

```python
data=pd.read_excel(r"D:\Travancore Analytics\Version 3\Datasets\Augmented_dataset 3.xlsx")
```

```python
data1 = data.dropna()
feature = data1.iloc[:,1:]
label = data1.iloc[:,0]
print(label)
```

```python
le=LabelEncoder()
label1=le.fit_transform(label)
print(label1)
print(label.unique())
```

```python
total_samples = len(feature)*100
num_categories = len(label.unique())
samples_per_category = total_samples // num_categories
print(samples_per_category)
```

```python
smote = SMOTE(sampling_strategy={label: samples_per_category for label in range(num_categories)}, random_state=42)
X_resampled, y_resampled = smote.fit_resample(feature, label1)
```

```python
unique=le.classes_
unique_labels = unique.tolist()
print(type(unique_labels))
print(unique_labels)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_resampled,y_resampled,test_size=0.2,random_state=42)
```

```python
model = xgb.XGBClassifier(objective='multi:softmax', num_class=num_categories,eval_metric='mlogloss')
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python

with open(r'D:\Travancore Analytics\Version 3\Models\xgboost.pkl', 'wb') as f:
    pickle.dump(model, f)
```
